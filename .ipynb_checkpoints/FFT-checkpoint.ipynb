```python
import numpy as np
import matplotlib.pyplot as plt

def plot_signal_and_spectrum(time_domain, signal, frequency_domain, spectrum, title):
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 1, 1)
    plt.plot(time_domain, signal)
    plt.title("Signal in Time Domain")
    plt.xlabel("Time (s)")
    plt.grid(True)

    plt.subplot(2, 1, 2)
    plt.plot(frequency_domain, np.abs(spectrum))
    plt.title("Spectrum in Frequency Domain")
    plt.xlabel("Frequency (Hz)")
    plt.grid(True)

    plt.tight_layout()
    plt.suptitle(title, fontsize=16)
    plt.show()

def fourier_transform(signal, sample_rate):
    # Compute the Discrete Fourier Transform (DFT)
    spectrum = np.fft.fft(signal)
    
    # Compute the corresponding frequencies for the spectrum
    frequencies = np.fft.fftfreq(len(signal), 1 / sample_rate)

    return frequencies, spectrum

# Example 1: Simple Sine Wave
sample_rate = 1000  # Sampling rate (samples per second)
duration = 1  # Duration of the signal in seconds
frequency = 5  # Frequency of the sine wave in Hz

time_domain = np.linspace(0, duration, sample_rate * duration)
signal = np.sin(2 * np.pi * frequency * time_domain)


frequencies, spectrum = fourier_transform(signal, sample_rate)
plot_signal_and_spectrum(time_domain, signal, frequencies, spectrum, "Simple Sine Wave")

# Example 2: Sum of Two Sine Waves
frequency1 = 10  # Frequency of the first sine wave in Hz
frequency2 = 30  # Frequency of the second sine wave in Hz

signal2 = np.sin(2 * np.pi * frequency1 * time_domain) + np.sin(2 * np.pi * frequency2 * time_domain)

frequencies, spectrum = fourier_transform(signal2, sample_rate)
plot_signal_and_spectrum(time_domain, signal2, frequencies, spectrum, "Sum of Two Sine Waves")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

def generate_signal(sample_rate, duration, frequencies, amplitudes):
    time_domain = np.linspace(0, duration, int(sample_rate * duration))
    signal = np.zeros_like(time_domain)
    for freq, amp in zip(frequencies, amplitudes):
        signal += amp * np.sin(2 * np.pi * freq * time_domain)
    return time_domain, signal

def plot_signal_and_spectrum(time_domain, signal, frequency_domain, spectrum, title):
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 1, 1)
    plt.plot(time_domain, signal)
    plt.title("Signal in Time Domain")
    plt.xlabel("Time (s)")
    plt.grid(True)

    plt.subplot(2, 1, 2)
    plt.plot(frequency_domain, np.abs(spectrum))
    plt.title("Spectrum in Frequency Domain")
    plt.xlabel("Frequency (Hz)")
    plt.grid(True)

    plt.tight_layout()
    plt.suptitle(title, fontsize=16)
    plt.show()

def fourier_transform(signal, sample_rate):
    # Compute the Discrete Fourier Transform (DFT)
    spectrum = fft(signal)

    # Compute the corresponding frequencies for the spectrum
    frequencies = fftfreq(len(signal), 1 / sample_rate)

    return frequencies, spectrum

# Example 1: Simple Sine Wave
sample_rate = 1000  # Sampling rate (samples per second)
duration = 1  # Duration of the signal in seconds
frequency = 5  # Frequency of the sine wave in Hz

time_domain, signal = generate_signal(sample_rate, duration, [frequency], [1])
frequencies, spectrum = fourier_transform(signal, sample_rate)
plot_signal_and_spectrum(time_domain, signal, frequencies, spectrum, "Simple Sine Wave")

# Example 2: Sum of Two Sine Waves
frequency1 = 10  # Frequency of the first sine wave in Hz
frequency2 = 30  # Frequency of the second sine wave in Hz

time_domain, signal = generate_signal(sample_rate, duration, [frequency1, frequency2], [1, 0.5])
frequencies, spectrum = fourier_transform(signal, sample_rate)
plot_signal_and_spectrum(time_domain, signal, frequencies, spectrum, "Sum of Two Sine Waves")
 
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_signal(sample_rate, duration, frequencies, amplitudes):
    time_domain = np.linspace(0, duration, int(sample_rate * duration))
    signal = np.zeros_like(time_domain)
    for freq, amp in zip(frequencies, amplitudes):
        signal += amp * np.sin(2 * np.pi * freq * time_domain)
    return time_domain, signal

def moving_window_fourier(signal, window_size):
    num_windows = len(signal) // window_size
    spectra = []

    for i in range(num_windows):
        window_start = i * window_size
        window_end = window_start + window_size
        window = signal[window_start:window_end]

        # Compute the Discrete Fourier Transform (DFT)
        spectrum = np.fft.fft(window)

        # Compute the corresponding frequencies for the spectrum
        frequencies = np.fft.fftfreq(window_size, 1 / sample_rate)

        spectra.append((frequencies, spectrum))

    return spectra

# Example: Real-time Moving Window Fourier Transform
sample_rate = 1000  # Sampling rate (samples per second)
duration = 10  # Duration of the signal in seconds
window_size = 100  # Size of the moving window

# Frequencies and amplitudes of various components in the signal
frequencies = [5, 15, 30, 50, 75]  # Frequencies in Hz
amplitudes = [1, 0.8, 0.5, 0.3, 0.2]

time_domain, signal = generate_signal(sample_rate, duration, frequencies, amplitudes)

# Process the signal in real-time using the moving window approach
spectra = moving_window_fourier(signal, window_size)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(time_domain, signal, label="Signal")
for frequencies, spectrum in spectra:
    plt.plot(frequencies, np.abs(spectrum), label="Window Spectrum")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Real-time Moving Window Fourier Transform")
plt.legend()
plt.grid(True)
plt.show()
```
