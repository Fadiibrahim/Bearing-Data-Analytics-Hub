```python
# %matplotlib notebook
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import mplcursors
# Load the .mat file
Normal_load_0        = scipy.io.loadmat('Normal_load_0.mat')
Inner_Race_7_load_0  = scipy.io.loadmat('Inner_Race_7_load_0.mat')
Inner_Race_14_load_0 = scipy.io.loadmat('Inner_Race_14_load_0.mat')
Inner_Race_21_load_0 = scipy.io.loadmat('Inner_Race_21_load_0.mat')
Inner_Race_28_load_0 = scipy.io.loadmat('Inner_Race_28_load_0.mat')


Normal_load_1        = scipy.io.loadmat('Normal_load_1.mat')
Inner_Race_7_load_1  = scipy.io.loadmat('Inner_Race_7_load_1.mat')
Inner_Race_14_load_1 = scipy.io.loadmat('Inner_Race_14_load_1.mat')
Inner_Race_21_load_1 = scipy.io.loadmat('Inner_Race_21_load_1.mat')
Inner_Race_28_load_1 = scipy.io.loadmat('Inner_Race_28_load_1.mat')


Normal_load_2        = scipy.io.loadmat('Normal_load_2.mat')
Inner_Race_7_load_2  = scipy.io.loadmat('Inner_Race_7_load_2.mat')
Inner_Race_14_load_2 = scipy.io.loadmat('Inner_Race_14_load_2.mat')
Inner_Race_21_load_2 = scipy.io.loadmat('Inner_Race_21_load_2.mat')
Inner_Race_28_load_2 = scipy.io.loadmat('Inner_Race_28_load_2.mat')


Normal_load_3        = scipy.io.loadmat('Normal_load_3.mat')
Inner_Race_7_load_3  = scipy.io.loadmat('Inner_Race_7_load_3.mat')
Inner_Race_14_load_3 = scipy.io.loadmat('Inner_Race_14_load_3.mat')
Inner_Race_21_load_3 = scipy.io.loadmat('Inner_Race_21_load_3.mat')
Inner_Race_28_load_3 = scipy.io.loadmat('Inner_Race_28_load_3.mat')


# Access the variables in the .mat file
# The variables are stored as keys in a dictionary
# For example, if your .mat file contains a variable 'data', you can access it as follows:
#data = mat_file['data']
S_n_L0  =  Normal_load_0['X097_DE_time']
S_7_L0  =  Inner_Race_7_load_0['X105_DE_time']
S_14_L0 =  Inner_Race_14_load_0['X169_DE_time']
S_21_L0 =  Inner_Race_21_load_0['X209_DE_time'] 
S_28_L0 =  Inner_Race_28_load_0['X056_DE_time'] 



S_n_L1  =      Normal_load_1['X100_DE_time']
S_7_L1  =  Inner_Race_7_load_1['X108_DE_time']
S_14_L1 =  Inner_Race_14_load_1['X172_DE_time']
S_21_L1 =  Inner_Race_21_load_1['X212_DE_time'] 
S_28_L1 =  Inner_Race_28_load_1['X059_DE_time'] 

S_n_L2  =      Normal_load_2['X098_DE_time']
S_7_L2  =  Inner_Race_7_load_2['X107_DE_time']
S_14_L2 =  Inner_Race_14_load_2['X171_DE_time']
S_21_L2 =  Inner_Race_21_load_2['X211_DE_time'] 
S_28_L2 =  Inner_Race_28_load_2['X058_DE_time']

S_n_L3  =      Normal_load_3['X100_DE_time']
S_7_L3  =  Inner_Race_7_load_3['X108_DE_time']
S_14_L3 =  Inner_Race_14_load_3['X172_DE_time']
S_21_L3 =  Inner_Race_21_load_3['X212_DE_time'] 
S_28_L3 =  Inner_Race_28_load_3['X059_DE_time'] 
```

```python
import matplotlib.pyplot as plt

Vars_cat_load_0 = ['S_n_L0', 'S_7_L0', 'S_14_L0', 'S_21_L0', 'S_28_L0']
Vars_cat_load_1 = ['S_n_L1', 'S_7_L1', 'S_14_L1', 'S_21_L1', 'S_28_L1']
Vars_cat_load_2 = ['S_n_L2', 'S_7_L2', 'S_14_L2', 'S_21_L2', 'S_28_L2']
Vars_cat_load_3 = ['S_n_L3', 'S_7_L3', 'S_14_L3', 'S_21_L3', 'S_28_L3']
plt.figure(figsize=(20, 10))
for i in range(len(Vars_cat_load_1)):
    plt.subplot(3, 2, i+1)
    plt.plot(eval(Vars_cat_load_0[i]), label=Vars_cat_load_0[i])
    plt.plot(eval(Vars_cat_load_1[i]), label=Vars_cat_load_1[i])
    plt.plot(eval(Vars_cat_load_2[i]), label=Vars_cat_load_2[i])
    plt.plot(eval(Vars_cat_load_3[i]), label=Vars_cat_load_3[i])
    plt.ylim(-4, 4)
    plt.legend()
    plt.xlabel('Time')
    plt.ylabel('Vib')
    
mplcursors.cursor(hover=True)
plt.show()
 
```

```python
plt.figure(figsize=(12, 6))
for i in range(len(Vars_cat_load_1)):
    plt.subplot(3, 2, i+1)
    plt.hist(eval(Vars_cat_load_3[i]), bins=50, label=Vars_cat_load_3[i])       
    plt.hist(eval(Vars_cat_load_2[i]), bins=50, label=Vars_cat_load_2[i])  
    plt.hist(eval(Vars_cat_load_1[i]), bins=50, label=Vars_cat_load_1[i])      
    plt.hist(eval(Vars_cat_load_0[i]), bins=50, label=Vars_cat_load_0[i])   
    plt.legend()
    plt.xlabel('Vib')
    plt.ylabel('freq')    
    plt.ylim(0, 40000)
    
    
    

        
```

```python
Var_laod_0_vec = []
Var_laod_1_vec = []
Var_laod_2_vec = []
Var_laod_3_vec = []
for i in range(len(Vars_cat_load_0)):
    Var_laod_0 = np.var(eval(Vars_cat_load_0[i]), ddof=0)
    Var_laod_1 = np.var(eval(Vars_cat_load_1[i]), ddof=0)
    Var_laod_2 = np.var(eval(Vars_cat_load_2[i]), ddof=0)
    Var_laod_3 = np.var(eval(Vars_cat_load_3[i]), ddof=0)
    Var_laod_0_vec.append(Var_laod_0)
    Var_laod_1_vec.append(Var_laod_1)
    Var_laod_2_vec.append(Var_laod_2)
    Var_laod_3_vec.append(Var_laod_3)

    

# Sample data
values = [10, 25, 15, 30]
cat = ['S_n', 'S_7', 'S_14', 'S_21', 'S_28']

plt.figure(figsize=(12, 6))
plt.bar(cat, Var_laod_0_vec, color='m', edgecolor='black', width=0.6,  label='load_0')
plt.bar(cat, Var_laod_1_vec, color='g', edgecolor='black', width=0.6,  label='load_1')
plt.bar(cat, Var_laod_2_vec, color='red', edgecolor='black', width=0.6,  label='load_2')
plt.bar(cat, Var_laod_3_vec, color='blue', edgecolor='black', width=0.6, label='load_3')
plt.legend()
# Add labels and title
plt.ylabel('Values')
plt.title('Bar Plot')

# Show the plot
plt.show()
  
```

```python
### BEGIN SOLUTION
# First we have to reshape the data so there is 
# only a single measurement in each column

plot_data = (Inner_Race_DE_df 
             .set_index('Load')
             .stack()
             .to_frame()
             .reset_index()
             .rename(columns={0:'size', 'level_1':'measurement'})
            )

plot_data 
### END SOLUTION
```

```python
### BEGIN SOLUTION
# Now plot the dataframe from above using Seaborn

sns.set_style('white')
sns.set_context('notebook')
sns.set_palette('dark')

f = plt.figure(figsize=(6,4))
sns.boxplot(x='measurement', y='size', 
            hue='Load', data=plot_data);
### END SOLUTION
```
