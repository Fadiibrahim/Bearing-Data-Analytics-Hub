```python
import matplotlib.pyplot as plt

Vars_cat_load_0 = ['N_L0', 'B_7_L0', 'B_14_L0', 'B_21_L0', 'B_28_L0']
Vars_cat_load_1 = ['N_L1', 'B_7_L1', 'B_14_L1', 'B_21_L1', 'B_28_L1']
Vars_cat_load_2 = ['N_L2', 'B_7_L2', 'B_14_L2', 'B_21_L2', 'B_28_L2']
Vars_cat_load_3 = ['N_L3', 'B_7_L3', 'B_14_L3', 'B_21_L3', 'B_28_L3']
plt.figure(figsize=(20, 10))
for i in range(len(Vars_cat_load_1)):
    plt.subplot(3, 2, i+1)
    plt.plot(eval(Vars_cat_load_0[i]), label=Vars_cat_load_0[i])
    plt.plot(eval(Vars_cat_load_1[i]), label=Vars_cat_load_1[i])
    plt.plot(eval(Vars_cat_load_2[i]), label=Vars_cat_load_2[i])
    plt.plot(eval(Vars_cat_load_3[i]), label=Vars_cat_load_3[i])
    plt.ylim(-10, 10)
    plt.legend()
    plt.xlabel('Time')
    plt.ylabel('Vib')
    
mplcursors.cursor(hover=True)
plt.show()
 
```

```python
plt.figure(figsize=(12, 6))
for i in range(len(Vars_cat_load_1)):
    plt.subplot(3, 2, i+1)
    plt.hist(eval(Vars_cat_load_3[i]), bins=50, label=Vars_cat_load_3[i])       
    plt.hist(eval(Vars_cat_load_2[i]), bins=50, label=Vars_cat_load_2[i])  
    plt.hist(eval(Vars_cat_load_1[i]), bins=50, label=Vars_cat_load_1[i])      
    plt.hist(eval(Vars_cat_load_0[i]), bins=50, label=Vars_cat_load_0[i])   
    plt.legend()
    plt.xlabel('Vib')
    plt.ylabel('freq')    
    plt.ylim(0, 40000)
    plt.xlim(-10, +10)
    
    
    

        
```

```python
Var_laod_0_vec = []
Var_laod_1_vec = []
Var_laod_2_vec = []
Var_laod_3_vec = []
for i in range(len(Vars_cat_load_0)):
    Var_laod_0 = np.var(eval(Vars_cat_load_0[i]), ddof=0)
    Var_laod_1 = np.var(eval(Vars_cat_load_1[i]), ddof=0)
    Var_laod_2 = np.var(eval(Vars_cat_load_2[i]), ddof=0)
    Var_laod_3 = np.var(eval(Vars_cat_load_3[i]), ddof=0)
    Var_laod_0_vec.append(Var_laod_0)
    Var_laod_1_vec.append(Var_laod_1)
    Var_laod_2_vec.append(Var_laod_2)
    Var_laod_3_vec.append(Var_laod_3)

    

# Sample data
values = [10, 25, 15, 30]
cat = ['B_n', 'B_7', 'B_14', 'B_21', 'B_28']

plt.figure(figsize=(12, 6))
plt.bar(cat, Var_laod_0_vec, color='m', edgecolor='black', width=0.6,  label='load_0')
plt.bar(cat, Var_laod_1_vec, color='g', edgecolor='black', width=0.6,  label='load_1')
plt.bar(cat, Var_laod_2_vec, color='red', edgecolor='black', width=0.6,  label='load_2')
plt.bar(cat, Var_laod_3_vec, color='blue', edgecolor='black', width=0.6, label='load_3')
plt.legend()
# Add labels and title
plt.ylabel('Values')
plt.title('Bar Plot')

# Show the plot
plt.show()
  
```
